```python
cara_sherlock = np.array(Image.open('sherlock.png'))
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.imshow(cara_sherlock)
plt.axis('off')
plt.show()
```

```python
sherlock_nube = WordCloud(background_color='white',
                         max_words = 2000,
                         mask = cara_sherlock,
                         stopwords = stopwords,
                         contour_width = 2,
                         contour_color='steelBlue')
sherlock_nube.generate(sherlock)

fig, ax = plt.subplots(figsize=(25,25))

ax.imshow(sherlock_nube, interpolation = 'bilinear')
ax.axis('off')
sherlock_nube.to_file('imagensherlock.png')
plt.show()
```
